```python
import numpy as np
size = 10000
mu, sigma = 0, 0.1

data = np.random.normal(mu, sigma, size)

import matplotlib.pyplot as plt
binsNum = 100
plt.hist(data, binsNum)
plt.show()
```

```python
for binsNum in range(5,100,5):
    plt.figure()
    plt.hist(data, binsNum)
    plt.title("Bins number: %d"%binsNum)
    plt.show()
```

```python
for mu in np.arange(-1.0, 1.0, 0.25):
    plt.figure()
    for sigma in np.arange(0.0, 50.0, 5.0):
        data = np.random.normal(mu, sigma, size)
        plt.hist(data, binsNum, alpha=0.25)
        plt.title("Mu: %.2f, Sigma: %.2f-%.2f"%(mu, 0.0, 45.0))
    plt.ylim(0, 400)
    plt.show()
```

```python
mu, sigma = 0, 0.1
data = np.random.normal(mu, sigma, size)
plt.figure()
plt.hist(data, 80, cumulative=1)
plt.show()
```

```python
import scipy
mu, sigma = 0, 0.1
data = scipy.special.logit(np.random.normal(mu, sigma, size))
plt.figure()
plt.hist(data, 80)
plt.show()
```

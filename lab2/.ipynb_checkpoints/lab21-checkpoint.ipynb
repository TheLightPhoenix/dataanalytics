## Normal distribution

```python
import numpy as np
size = 10000
mu, sigma = 0, 0.1

data = np.random.normal(mu, sigma, size)

import matplotlib.pyplot as plt
binsNum = 100
plt.hist(data, binsNum)
plt.show()
```

```python
for binsNum in range(5,100,5):
    plt.figure()
    plt.hist(data, binsNum)
    plt.title("Bins number: %d"%binsNum)
    plt.show()
```

With more bins we get better resolution and see more details. If bins number isn't enough we will get wrong idea of how data distribution looks like. More bins increases computational time. If we set too much bins it will take longer to compute and it won't give us any more information about data.

```python
for mu in np.arange(-1.0, 1.0, 0.25):
    plt.figure()
    for sigma in np.arange(0.0, 50.0, 5.0):
        data = np.random.normal(mu, sigma, size)
        plt.hist(data, binsNum, alpha=0.25)
        plt.title("Mu: %.2f, Sigma: %.2f-%.2f"%(mu, 0.0, 45.0))
    plt.ylim(0, 400)
    plt.show()
```

Changing mu moves whole distibtution to the left or right (changes mean). Increasing sigma makes distribution wider. Bigger value of sigma makes distribution less informative, smaller - informative.

```python
mu, sigma = 0, 0.1
data = np.random.normal(mu, sigma, size)
plt.figure()
plt.hist(data, 80, cumulative=1)
plt.show()
```

## Inverse logit

```python
import scipy
from scipy import special
mu, sigma = 0, 0.1
data = scipy.special.logit(np.random.normal(mu, sigma, size))
plt.figure()
plt.hist(data, 80)
plt.show()
```

```python
for mu in np.arange(0.0, 1.0, 0.25):
    plt.figure()
    for sigma in np.arange(0.1, 1.0, 0.1):
        data = scipy.special.logit(np.random.normal(mu, sigma, size))
        plt.hist(data, binsNum, alpha=0.25)
        plt.title("Mu: %.2f, Sigma: %.2f-%.2f"%(mu, 0.0, 45.0))
    plt.legend(np.arange(0.1, 1.0, 0.1))
    plt.ylim(0, 800)
    plt.show()
```

Now changing sigma makes distribution shorter. After transformation non-informative distributions became informative.

```python
for mu in np.arange(0.0, 0.75, 0.25):
    plt.figure()
    for sigma in np.arange(0.1, 1.0, 0.1):
        data = scipy.special.logit(np.random.normal(mu, sigma, size))
        plt.hist(data, binsNum, alpha=0.25, cumulative=1)
        plt.title("Mu: %.2f, Sigma: %.2f-%.2f"%(mu, 0.0, 45.0))
    plt.legend(np.arange(0.1, 1.0, 0.1))
    plt.show()
```

Increasing sigma makes cdf plot longer

## Poisson distribution

```python
import pandas as pd

dataDict = {}
for x in range(0, 10, 1):
    dataDict[x] = np.random.poisson(x, size)
dataPd = pd.DataFrame(dataDict)
dataMelt = pd.melt(dataPd)
import seaborn as sns
grid = sns.FacetGrid(data=dataMelt, col="variable", col_wrap=3)
grid.map(plt.hist, "value", bins=range(0,20,1))
```

Increasing lambda value makes distribution less concentrated and move from 0 value.

## Beta distribution

```python
dataDict = {}
for a in np.arange(0.5, 2, 0.5):
    for b in np.arange(0.5, 2, 0.5):
        dataDict['a=' + str(a) + ' b=' + str(b)] = np.random.beta(a,b, size)
dataPd = pd.DataFrame(dataDict)
dataMelt = pd.melt(dataPd)

import seaborn as sns
grid = sns.FacetGrid(data=dataMelt, col="variable", col_wrap=3)
grid.map(plt.hist, "value", bins=50)
```

Increasing one value makes distribution be more concentrated in one side (alfa - left, beta - right). For values 1,1 this distribution looks uniform, for values greater than one distribution is starting to rise in the middle.

### Location dispersion parametrization

```python
import pandas as pd
dataDict = {}
for u in np.arange(0.1, 1.0, 0.2):
    for y in np.arange(0.1, 1.0, 0.2):
        dataDict['u=' + str(round(u,2)) + ' y=' + str(round(y,2))] = np.random.beta(u*(1-y)/y, (1-u)*(1-y)/y, size)
dataPd = pd.DataFrame(dataDict)
dataMelt = pd.melt(dataPd)

import seaborn as sns
grid = sns.FacetGrid(data=dataMelt, col="variable", col_wrap=5)
grid.map(plt.hist, "value", bins=20)
```

Changing u moves distribution's concetration (mean). y changes dispersion (how much distribution is concentrated)

### Mean-population size parametrization

```python

dataDict = {}
for u in np.arange(0.1, 1.0, 0.2):
    for v in np.arange(0.1, 1.0, 0.2):
        dataDict['u=' + str(round(u,2)) + ' v=' + str(round(v,2))] = np.random.beta(u*v, (1-u)*v, size)
dataPd = pd.DataFrame(dataDict)
dataMelt = pd.melt(dataPd)

import seaborn as sns
grid = sns.FacetGrid(data=dataMelt, col="variable", col_wrap=5)
grid.map(plt.hist, "value", bins=20)
```

u changes proportion of values close to 0 and 1. v changes concentration

## Lognormal distribution

```python

dataDict = {}
for mu in np.arange(0.0, 1.0, 0.2):
    for sigma in np.arange(0.1, 0.6, 0.1):
        dataDict['mu=' + str(round(mu,2)) + ' sigma=' + str(round(sigma,2))] = np.random.lognormal(mu, sigma, size)
dataPd = pd.DataFrame(dataDict)
dataMelt = pd.melt(dataPd)

import seaborn as sns
grid = sns.FacetGrid(data=dataMelt, col="variable", col_wrap=5)
grid.map(plt.hist, "value", bins=20)
```

Sigma changes concentration, mu changes mean - moves distribution.
